Spelling Correction

```python
from tqdm import tqdm
from spellchecker import SpellChecker

def spell_check(words):
    spell = SpellChecker()
    corrected_words = []
    # ä½¿ç”¨tqdmä¾†é¡¯ç¤ºé€²åº¦æ¢
    for word in tqdm(words, desc="Spelling Check"):
        corrected_word = spell.correction(word)
        corrected_words.append(corrected_word)
    return corrected_words

# å° "Tweet" æ¬„ä½ä¸­çš„æ¯å€‹åˆ—è¡¨é€²è¡Œæ‹¼å¯«æª¢æŸ¥
textdata_df["Tweet"] = textdata_df["Tweet"].apply(spell_check)

# è¼¸å‡ºä¿®æ­£å¾Œçš„æ–‡æœ¬
print(textdata_df["Tweet"].head())

```

Removing Stopwords

```python
import nltk
from nltk.corpus import stopwords

textdata_df = pd.read_csv('E:\\code\information_retrieve\\result.tsv', sep='\t')
textdata_df
textdata_df["Tweet"] = textdata_df["Tweet"].apply(lambda tweet: [word if isinstance(word, str) else 'None' for word in tweet])
# ä¸‹è¼‰åœç”¨è©è³‡æº
nltk.download('stopwords')

# åŠ è¼‰è‹±æ–‡åœç”¨è©åˆ—è¡¨
stop_words = set(stopwords.words('english'))

# å®šç¾©å‡½å¼ä¾†ç§»é™¤åœç”¨è©
def remove_stopwords(series):
    # å»é™¤åœç”¨è©ä¸¦çµ„åˆå–®è©
    filtered_tweets = []
    for tweet in series:
        if not tweet:
            filtered_tweets.append([])
            continue
        # å°‡åˆ—è¡¨ä¸­çš„å­—å…ƒçµ„åˆæˆå­—ä¸²
        tweet = ''.join(tweet)
        # å»é™¤åœç”¨è©ä¸¦çµ„åˆå–®è©
        filtered_tweet = [word for word in tweet.split() if word.lower() not in stop_words]
        filtered_tweets.append(filtered_tweet)
    return filtered_tweets

# å° "Tweet" æ¬„ä½ä¸­çš„æ¯å€‹åˆ—è¡¨å»é™¤åœç”¨è©
textdata_df["Tweet"] = remove_stopwords(textdata_df["Tweet"])

# è¼¸å‡ºå»é™¤åœç”¨è©å¾Œçš„æ–‡æœ¬
print(textdata_df["Tweet"].head())
```

Removing Punctuations

```python
from nltk.tokenize import RegexpTokenizer

def remove_punct(text):
    tokenizer = RegexpTokenizer(r"\w+")
    lst = tokenizer.tokenize(' '.join(text))
    return lst

# å°‡è™•ç†å¾Œçš„æ¨æ–‡å­˜å›åˆ°åŸå§‹çš„ DataFrame ä¸­
textdata_df["Tweet"] = textdata_df["Tweet"].apply(remove_punct)

# æ¸¬è©¦
print(textdata_df["Tweet"].head())
```

Removing Frequent Words

```python
from nltk.probability import FreqDist
from nltk.tokenize import word_tokenize

# å°‡æ‰€æœ‰æ¨æ–‡çµ„åˆæˆä¸€å€‹å­—ç¬¦ä¸²
all_tweets = ' '.join(textdata_df["Tweet"].sum())

# åˆ†è©
tokens = word_tokenize(all_tweets)

# å‰µå»ºè©é »åˆ†ä½ˆ
freq_dist = FreqDist(tokens)

# æ‰¾åˆ°æœ€å¸¸è¦‹çš„å–®è©ï¼Œé€™è£¡å‡è¨­æ‚¨å¸Œæœ›åˆªé™¤å‰ n å€‹æœ€å¸¸è¦‹çš„å–®è©
n = 10
most_common_words = freq_dist.most_common(n)

# åˆ—å°æœ€å¸¸è¦‹çš„å–®è©
print("Most common words:")
for word, freq in most_common_words:
    print(f"{word}: {freq}")

# å¾æ¯å€‹æ¨æ–‡ä¸­åˆªé™¤æœ€å¸¸è¦‹çš„å–®è©
def remove_common_words(tweet):
    common_words = ["None"]  
    return [word for word in tweet if word not in common_words]

textdata_df["Tweet"] = textdata_df["Tweet"].apply(remove_common_words)

# è¼¸å‡ºè™•ç†å¾Œçš„æ–‡æœ¬
print(textdata_df["Tweet"].head())


textdata_df["Tweet"] = textdata_df["Tweet"].apply(remove_common_words)

# è¼¸å‡ºè™•ç†å¾Œçš„æ–‡æœ¬
print(textdata_df["Tweet"].head())
```

Lemmatization

```python
import nltk
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
```

```python
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
from nltk.tokenize import word_tokenize

def lemmatization(text):
    lemmatizer = WordNetLemmatizer()
    result = []
    for token, tag in pos_tag(text):
        pos = tag[0].lower()
        if pos not in ['a', 'r', 'n', 'v']:
            pos = 'n'
        result.append(lemmatizer.lemmatize(token, pos))
    return result

# å°‡æ¨æ–‡åˆ—è¡¨è½‰æ›ç‚ºå–®å€‹æ–‡æœ¬å­—ç¬¦ä¸²ï¼Œå†é€²è¡Œè©å½™åŒ–è™•ç†
textdata_df["Tweet"] = textdata_df["Tweet"].apply(lambda tweet: lemmatization(tweet))

# æ¸¬è©¦
print(textdata_df["Tweet"].head())
```

```python
from nltk.stem import PorterStemmer

def stemming(text):
    porter = PorterStemmer()
    result = []
    for word in text:
        result.append(porter.stem(word))
    return result

# å°‡æ¨æ–‡åˆ—è¡¨è½‰æ›ç‚ºå–®å€‹æ–‡æœ¬å­—ç¬¦ä¸²ï¼Œå†é€²è¡Œè©å¹¹æå–è™•ç†
textdata_df["Tweet"] = textdata_df["Tweet"].apply(lambda tweet: stemming(tweet))

# æ¸¬è©¦
print(textdata_df["Tweet"].head())
```

 Removal of Tags

```python
import re

def remove_tag(text):
    text = ' '.join(text)
    html_pattern = re.compile('<.*?>')
    return html_pattern.sub(r'', text)

# å°‡æ¨æ–‡åˆ—è¡¨è½‰æ›ç‚ºå–®å€‹æ–‡æœ¬å­—ç¬¦ä¸²ï¼Œå†é€²è¡Œæ¨™ç±¤ç§»é™¤è™•ç†
textdata_df["Tweet"] = textdata_df["Tweet"].apply(lambda tweet: remove_tag(tweet))

# æ¸¬è©¦
print(textdata_df["Tweet"].head())
```

Removal of URLs 

```python
import re

def remove_urls(text):
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return url_pattern.sub(r'', text)

# å°‡æ¨æ–‡åˆ—è¡¨è½‰æ›ç‚ºå–®å€‹æ–‡æœ¬å­—ç¬¦ä¸²ï¼Œå†é€²è¡Œ URL ç§»é™¤è™•ç†
textdata_df["Tweet"] = textdata_df["Tweet"].apply(lambda tweet: remove_urls(tweet))

# æ¸¬è©¦
print(textdata_df["Tweet"].head())
```
