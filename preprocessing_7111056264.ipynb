```python
def remove_whitespace(text):
    return " ".join(text.split())

# Â∞áÂáΩÂºèÊáâÁî®Âà∞"Tweet"Ê¨Ñ‰Ωç
textdata_df["Tweet"] = textdata_df["Tweet"].apply(remove_whitespace)

# Ê™¢Êü•"Tweet"Ê¨Ñ‰ΩçÊòØÂê¶Â∑≤Á∂ìÁßªÈô§‰∫ÜÁ©∫ÁôΩ
print(textdata_df["Tweet"])
```

Spelling Correction

```python
from tqdm import tqdm
from spellchecker import SpellChecker

def spell_check(words):
    spell = SpellChecker()
    corrected_words = []
    # ‰ΩøÁî®tqdm‰æÜÈ°ØÁ§∫ÈÄ≤Â∫¶Ê¢ù
    for word in tqdm(words, desc="Spelling Check"):
        corrected_word = spell.correction(word)
        corrected_words.append(corrected_word)
    return corrected_words

# Â∞ç "Tweet" Ê¨Ñ‰Ωç‰∏≠ÁöÑÊØèÂÄãÂàóË°®ÈÄ≤Ë°åÊãºÂØ´Ê™¢Êü•
textdata_df["Tweet"] = textdata_df["Tweet"].apply(spell_check)

# Ëº∏Âá∫‰øÆÊ≠£ÂæåÁöÑÊñáÊú¨
print(textdata_df["Tweet"].head())

```

Removing Stopwords

```python
import nltk
from nltk.corpus import stopwords

textdata_df = pd.read_csv('E:\\code\information_retrieve\\result.tsv', sep='\t')
textdata_df
textdata_df["Tweet"] = textdata_df["Tweet"].apply(lambda tweet: [word if isinstance(word, str) else 'None' for word in tweet])
# ‰∏ãËºâÂÅúÁî®Ë©ûË≥áÊ∫ê
nltk.download('stopwords')

# Âä†ËºâËã±ÊñáÂÅúÁî®Ë©ûÂàóË°®
stop_words = set(stopwords.words('english'))

# ÂÆöÁæ©ÂáΩÂºè‰æÜÁßªÈô§ÂÅúÁî®Ë©û
def remove_stopwords(series):
    # ÂéªÈô§ÂÅúÁî®Ë©û‰∏¶ÁµÑÂêàÂñÆË©û
    filtered_tweets = []
    for tweet in series:
        if not tweet:
            filtered_tweets.append([])
            continue
        # Â∞áÂàóË°®‰∏≠ÁöÑÂ≠óÂÖÉÁµÑÂêàÊàêÂ≠ó‰∏≤
        tweet = ''.join(tweet)
        # ÂéªÈô§ÂÅúÁî®Ë©û‰∏¶ÁµÑÂêàÂñÆË©û
        filtered_tweet = [word for word in tweet.split() if word.lower() not in stop_words]
        filtered_tweets.append(filtered_tweet)
    return filtered_tweets

# Â∞ç "Tweet" Ê¨Ñ‰Ωç‰∏≠ÁöÑÊØèÂÄãÂàóË°®ÂéªÈô§ÂÅúÁî®Ë©û
textdata_df["Tweet"] = remove_stopwords(textdata_df["Tweet"])

# Ëº∏Âá∫ÂéªÈô§ÂÅúÁî®Ë©ûÂæåÁöÑÊñáÊú¨
print(textdata_df["Tweet"].head())
```

Removing Punctuations

```python
from nltk.tokenize import RegexpTokenizer

def remove_punct(text):
    tokenizer = RegexpTokenizer(r"\w+")
    lst = tokenizer.tokenize(' '.join(text))
    return lst

# Â∞áËôïÁêÜÂæåÁöÑÊé®ÊñáÂ≠òÂõûÂà∞ÂéüÂßãÁöÑ DataFrame ‰∏≠
textdata_df["Tweet"] = textdata_df["Tweet"].apply(remove_punct)

# Ê∏¨Ë©¶
print(textdata_df["Tweet"].head())
```

Removing Frequent Words

```python
from nltk.probability import FreqDist
from nltk.tokenize import word_tokenize

# Â∞áÊâÄÊúâÊé®ÊñáÁµÑÂêàÊàê‰∏ÄÂÄãÂ≠óÁ¨¶‰∏≤
all_tweets = ' '.join(textdata_df["Tweet"].sum())

# ÂàÜË©û
tokens = word_tokenize(all_tweets)

# ÂâµÂª∫Ë©ûÈ†ªÂàÜ‰Ωà
freq_dist = FreqDist(tokens)

# ÊâæÂà∞ÊúÄÂ∏∏Ë¶ãÁöÑÂñÆË©ûÔºåÈÄôË£°ÂÅáË®≠ÊÇ®Â∏åÊúõÂà™Èô§Ââç n ÂÄãÊúÄÂ∏∏Ë¶ãÁöÑÂñÆË©û
n = 10
most_common_words = freq_dist.most_common(n)

# ÂàóÂç∞ÊúÄÂ∏∏Ë¶ãÁöÑÂñÆË©û
print("Most common words:")
for word, freq in most_common_words:
    print(f"{word}: {freq}")

# ÂæûÊØèÂÄãÊé®Êñá‰∏≠Âà™Èô§ÊúÄÂ∏∏Ë¶ãÁöÑÂñÆË©û
def remove_common_words(tweet):
    common_words = ["None"]  
    return [word for word in tweet if word not in common_words]

textdata_df["Tweet"] = textdata_df["Tweet"].apply(remove_common_words)

# Ëº∏Âá∫ËôïÁêÜÂæåÁöÑÊñáÊú¨
print(textdata_df["Tweet"].head())


textdata_df["Tweet"] = textdata_df["Tweet"].apply(remove_common_words)

# Ëº∏Âá∫ËôïÁêÜÂæåÁöÑÊñáÊú¨
print(textdata_df["Tweet"].head())
```

Lemmatization

```python
import nltk
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
```

```python
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
from nltk.tokenize import word_tokenize

def lemmatization(text):
    lemmatizer = WordNetLemmatizer()
    result = []
    for token, tag in pos_tag(text):
        pos = tag[0].lower()
        if pos not in ['a', 'r', 'n', 'v']:
            pos = 'n'
        result.append(lemmatizer.lemmatize(token, pos))
    return result

# Â∞áÊé®ÊñáÂàóË°®ËΩâÊèõÁÇ∫ÂñÆÂÄãÊñáÊú¨Â≠óÁ¨¶‰∏≤ÔºåÂÜçÈÄ≤Ë°åË©ûÂΩôÂåñËôïÁêÜ
textdata_df["Tweet"] = textdata_df["Tweet"].apply(lambda tweet: lemmatization(tweet))

# Ê∏¨Ë©¶
print(textdata_df["Tweet"].head())
```

```python
from nltk.stem import PorterStemmer

def stemming(text):
    porter = PorterStemmer()
    result = []
    for word in text:
        result.append(porter.stem(word))
    return result

# Â∞áÊé®ÊñáÂàóË°®ËΩâÊèõÁÇ∫ÂñÆÂÄãÊñáÊú¨Â≠óÁ¨¶‰∏≤ÔºåÂÜçÈÄ≤Ë°åË©ûÂππÊèêÂèñËôïÁêÜ
textdata_df["Tweet"] = textdata_df["Tweet"].apply(lambda tweet: stemming(tweet))

# Ê∏¨Ë©¶
print(textdata_df["Tweet"].head())
```

 Removal of Tags

```python
import re

def remove_tag(text):
    text = ' '.join(text)
    html_pattern = re.compile('<.*?>')
    return html_pattern.sub(r'', text)

# Â∞áÊé®ÊñáÂàóË°®ËΩâÊèõÁÇ∫ÂñÆÂÄãÊñáÊú¨Â≠óÁ¨¶‰∏≤ÔºåÂÜçÈÄ≤Ë°åÊ®ôÁ±§ÁßªÈô§ËôïÁêÜ
textdata_df["Tweet"] = textdata_df["Tweet"].apply(lambda tweet: remove_tag(tweet))

# Ê∏¨Ë©¶
print(textdata_df["Tweet"].head())
```

Removal of URLs 

```python
import re

def remove_urls(text):
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return url_pattern.sub(r'', text)

# Â∞áÊé®ÊñáÂàóË°®ËΩâÊèõÁÇ∫ÂñÆÂÄãÊñáÊú¨Â≠óÁ¨¶‰∏≤ÔºåÂÜçÈÄ≤Ë°å URL ÁßªÈô§ËôïÁêÜ
textdata_df["Tweet"] = textdata_df["Tweet"].apply(lambda tweet: remove_urls(tweet))

# Ê∏¨Ë©¶
print(textdata_df["Tweet"].head())
```
